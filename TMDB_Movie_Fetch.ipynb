### Check for spaces


```python
df['title'] = df['title'].str.strip()
df['overview'] = df['overview'].str.strip()
```

### Check for outliers

```python
df[['popularity', 'vote_average', 'vote_count']].describe()
```

```python
import numpy as np


Q1 = df['popularity'].quantile(0.25)
Q3 = df['popularity'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


df['popularity'] = np.where(df['popularity'] > upper_bound, upper_bound,
                     np.where(df['popularity'] < lower_bound, lower_bound, df['popularity']))
```

```python
df[['popularity', 'vote_average', 'vote_count']].describe()
```

### Verify Date range


```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

df[df['release_date'] > pd.Timestamp.today()]
df[df['release_date'] < '1900-01-01']
```
